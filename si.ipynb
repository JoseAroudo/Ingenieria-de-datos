### Ahora vamos a extraer la informaciÃ³n de la temperatura, la probabilidad de lluvia y el mensaje de la condiciÃ³n del clima y almacenarlo en un Dataframe

```python
def get_forecast(response, i):
    date = response["forecast"]["forecastday"][0]["hour"][i]["time"].split()[0]
    time = int(response["forecast"]["forecastday"][0]["hour"][i]["time"].split()[1].split(":")[0])
    temp = response["forecast"]["forecastday"][0]["hour"][i]["temp_c"]
    prob_rain = response["forecast"]["forecastday"][0]["hour"][i]["chance_of_rain"]
    will_rain = response["forecast"]["forecastday"][0]["hour"][i]["will_it_rain"]
    
    message = response["forecast"]["forecastday"][0]["hour"][i]["condition"]["text"]
    
    return date, time, temp, prob_rain, will_rain, message
```

```python
data = []

for i in tqdm(range(len(response["forecast"]["forecastday"][0]["hour"]))):
    data.append(get_forecast(response, i))
```

```python
data[11]
```

```python
columns = ["Date", "Time", "Temperature", "Probability of Rain","Will Rain", "Message"]
df = pd.DataFrame(data, columns=columns)
df
```

```python
df_final= df[(df["Will Rain"]==1) & (df["Time"]>=6)]
df_final = df_final[["Time", "Message"]]
df_final.set_index("Time", inplace=True)
df_final
```
